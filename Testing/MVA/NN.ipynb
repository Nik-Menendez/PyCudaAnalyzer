```python
#Create model

inputs = Input(shape=(len(vars_test),), name = 'input')
#x = Dropout(.1, name = 'input_dropout_.1')(inputs)
x = Dense(32, name = 'hidden1')(inputs)
x = LeakyReLU(.1, name = 'LeakyReLU1_.1')(x)
x = Dense(64, name = 'hidden2')(x)
x = LeakyReLU(.1, name = 'LeakyReLU2_.1')(x)
x = Dense(32, name = 'hidden3')(x)
x = LeakyReLU(.1, name = 'LeakyReLU3_.1')(x)
#x = Dropout(.1, name = 'hidden2_dropout_.1')(x)
outputs = Dense(3, name = 'output', activation='softmax')(x)

model = Model(inputs=inputs, outputs=outputs)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])

# model = Sequential()
# model.add(Dense(64, input_dim=len(vars_test), activation='relu'))
# model.add(Dense(128, activation='relu'))
# model.add(Dense(4, activation='softmax'))
# # Compile model
# model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```
