```python
for m in masses:
    sample = 'WpTo3l_ZpM' + str(m)
    in_path = "/cmsuf/data/store/user/t2/users/nikmenendez/skimmed/signal/"+ sample + ".root"
    File = TFile(in_path,"READ")
    event = File.Get("passedEvents")
    nEntries = event.GetEntries()
    
    tot_events = 0
    low_cor_events = 0
    hig_cor_events = 0
    nn_cor_events = 0
    nna_cor_events = 0
    
    for e in tqdm(range(0, nEntries)):
        event.GetEntry(e)

        Lep1_mom = 0
        Lep2_mom = 0
        Lep3_mom = 0
        Lep1, Lep2, Lep3, Met = TLorentzVector(), TLorentzVector(), TLorentzVector(), TLorentzVector(),
        Met.SetPtEtaPhiM(event.met,0,event.met_phi,0)
        save_event = False

        if event.pTL3 > event.pTL1:

            Lep1_pt = event.pTL3#/event.massL3
            Lep1_eta = event.etaL3
            Lep1_phi = event.phiL3
            Lep1_iso = event.IsoL3
            Lep1_id = event.idL3
            Lep1_mom = event.MomIdL3
            Lep1.SetPtEtaPhiM(event.pTL3,event.etaL3,event.phiL3,event.massL3)

            Lep2_pt = event.pTL1#/event.massL1
            Lep2_eta = event.etaL1
            Lep2_phi = event.phiL1
            Lep2_iso = event.IsoL1
            Lep2_id = event.idL1
            Lep2_mom = event.MomIdL1
            Lep2.SetPtEtaPhiM(event.pTL1,event.etaL1,event.phiL1,event.massL1)

            Lep3_pt = event.pTL2#/event.massL2
            Lep3_eta = event.etaL2
            Lep3_phi = event.phiL2
            Lep3_iso = event.IsoL2      
            Lep3_id = event.idL2
            Lep3_mom = event.MomIdL2
            Lep3.SetPtEtaPhiM(event.pTL2,event.etaL2,event.phiL2,event.massL2)

        elif event.pTL3 > event.pTL2 and event.pTL1 > event.pTL3:

            Lep1_pt = event.pTL1#/event.massL1
            Lep1_eta = event.etaL1
            Lep1_phi = event.phiL1
            Lep1_iso = event.IsoL1
            Lep1_id = event.idL1
            Lep1_mom = event.MomIdL1
            Lep1.SetPtEtaPhiM(event.pTL1,event.etaL1,event.phiL1,event.massL1)

            Lep2_pt = event.pTL3#/event.massL3
            Lep2_eta = event.etaL3
            Lep2_phi = event.phiL3
            Lep2_iso = event.IsoL3
            Lep2_id = event.idL3
            Lep2_mom = event.MomIdL3
            Lep2.SetPtEtaPhiM(event.pTL3,event.etaL3,event.phiL3,event.massL3)

            Lep3_pt = event.pTL2#/event.massL2
            Lep3_eta = event.etaL2
            Lep3_phi = event.phiL2
            Lep3_iso = event.IsoL2    
            Lep3_id = event.idL2
            Lep3_mom = event.MomIdL2
            Lep3.SetPtEtaPhiM(event.pTL2,event.etaL2,event.phiL2,event.massL2)

        elif event.pTL3 < event.pTL2:

            Lep1_pt = event.pTL1#/event.massL1
            Lep1_eta = event.etaL1
            Lep1_phi = event.phiL1
            Lep1_iso = event.IsoL1
            Lep1_id = event.idL1
            Lep1_mom = event.MomIdL1
            Lep1.SetPtEtaPhiM(event.pTL1,event.etaL1,event.phiL1,event.massL1)

            Lep2_pt = event.pTL2#/event.massL2
            Lep2_eta = event.etaL2
            Lep2_phi = event.phiL2
            Lep2_iso = event.IsoL2
            Lep2_id = event.idL2
            Lep2_mom = event.MomIdL2
            Lep2.SetPtEtaPhiM(event.pTL2,event.etaL2,event.phiL2,event.massL2)

            Lep3_pt = event.pTL3#/event.massL3
            Lep3_eta = event.etaL3
            Lep3_phi = event.phiL3
            Lep3_iso = event.IsoL3
            Lep3_id = event.idL3
            Lep3_mom = event.MomIdL3
            Lep3.SetPtEtaPhiM(event.pTL3,event.etaL3,event.phiL3,event.massL3)

        met = event.met
        met_phi = event.met_phi

        dR12 = deltaR(Lep1_eta,Lep1_phi,Lep2_eta,Lep2_phi)
        dR13 = deltaR(Lep1_eta,Lep1_phi,Lep3_eta,Lep3_phi)
        dR23 = deltaR(Lep2_eta,Lep2_phi,Lep3_eta,Lep3_phi)

        if abs(Lep1_mom) == zid and abs(Lep2_mom) == zid:
            corr_slot = 1
            #CorrMass = (Lep1 + Lep2).M()
        elif abs(Lep1_mom) == zid and abs(Lep3_mom) == zid:
            corr_slot = 2
            #CorrMass = (Lep1 + Lep3).M()
        elif abs(Lep2_mom) == zid and abs(Lep3_mom) == zid:
            corr_slot = 3
            #CorrMass = (Lep2 + Lep3).M()
        else:
            corr_slot = 0
            #CorrMass = 0

        # ---------------- Low Mass Selection ----------------
        if Lep1_id + Lep2_id == 0:
            #P2 = Lep1 + Lep2
            if Lep1_id + Lep3_id == 0:
                low_guess_slot = 2
                #P1 = Lep1 + Lep3
            else:
                low_guess_slot = 3
                #P1 = Lep2 + Lep3
        elif Lep1_id + Lep3_id == 0:
            #P2 = Lep1 + Lep3
            if Lep1_id + Lep2_id == 0:
                low_guess_slot = 1
                #P1 = Lep1 + Lep2
            else:
                low_guess_slot = 3
                #P1 = Lep2 + Lep3
        # ----------------------------------------------------
                
        # ---------------- High Mass Selection ---------------
        if Lep1_id + Lep2_id == 0:
            #P1 = Lep1 + Lep2
            hig_guess_slot = 1
        elif Lep1_id + Lep3_id == 0:
            #P1 = Lep1 + Lep3
            hig_guess_slot = 2
        # ----------------------------------------------------
        
        # ------------- Neural Network Selection -------------
        Selector_vars = np.array((Lep1_pt,Lep1_eta,Lep1_phi,Lep1_iso,Lep1_id,
                                  Lep2_pt,Lep2_eta,Lep2_phi,Lep2_iso,Lep2_id,
                                  Lep3_pt,Lep3_eta,Lep3_phi,Lep3_iso,Lep3_id,
                                  met, met_phi, dR12, dR13, dR23))
        Selector_vars = normalize(Selector_vars)
        Selector_vars = Selector_vars.reshape(1,-1)
        prediction_0 = ZSelector.predict(Selector_vars)
        
        max_nn_guess = -1
        nn_guess_slot = -1
        for i in range(len(prediction_0[0])):
            if prediction_0[0][i] > max_nn_guess:
                max_nn_guess = prediction_0[0][i]
                nn_guess_slot = i
        # ----------------------------------------------------
        
        # ----------- Neural Network Alt Selection -----------
        if Lep1_id + Lep2_id == 0:
            M1 = (Lep1 + Lep2).M()
            
            Lep1pt = Lep1_pt/M1
            Lep1eta = Lep1_eta
            Lep1phi = Lep1_phi
            Lep1iso = Lep1_iso
            Lep1id = Lep1_id
            Lep1mom = Lep1_mom

            Lep2pt = Lep2_pt/M1
            Lep2eta = Lep2_eta
            Lep2phi = Lep2_phi
            Lep2iso = Lep2_iso
            Lep2id = Lep2_id
            Lep2mom = Lep2_mom
            
            dR = deltaR(Lep1eta,Lep1phi,Lep2eta,Lep2phi)
            
            Selector_alts = np.array((Lep1pt,Lep1eta,Lep1phi,Lep1iso,Lep1id,
                                      Lep2pt,Lep2eta,Lep2phi,Lep2iso,Lep2id,
                                      met, met_phi, dR))
            Selector_alts = normalize_alt(Selector_alts)
            Selector_alts = Selector_alts.reshape(1,-1)
            prediction_1 = ZSelector_alt.predict(Selector_alts)
            
            if Lep1_id + Lep3_id == 0:
                M2 = (Lep1 + Lep3).M()
                
                Lep1pt = Lep1_pt/M2
                Lep1eta = Lep1_eta
                Lep1phi = Lep1_phi
                Lep1iso = Lep1_iso
                Lep1id = Lep1_id
                Lep1mom = Lep1_mom

                Lep2pt = Lep3_pt/M2
                Lep2eta = Lep3_eta
                Lep2phi = Lep3_phi
                Lep2iso = Lep3_iso
                Lep2id = Lep3_id
                Lep2mom = Lep3_mom

                dR = deltaR(Lep1eta,Lep1phi,Lep2eta,Lep2phi)

                Selector_alts = np.array((Lep1pt,Lep1eta,Lep1phi,Lep1iso,Lep1id,
                                          Lep2pt,Lep2eta,Lep2phi,Lep2iso,Lep2id,
                                          met, met_phi, dR))
                Selector_alts = normalize_alt(Selector_alts)
                Selector_alts = Selector_alts.reshape(1,-1)
                prediction_2 = ZSelector_alt.predict(Selector_alts)
                
                if prediction_1[0] > prediction_2[0]:
                    nna_guess_slot = 1
                else:
                    nna_guess_slot = 2
                
            elif Lep2_id + Lep3_id == 0:
                M2 = (Lep2 + Lep3).M()
                
                Lep1pt = Lep2_pt/M2
                Lep1eta = Lep2_eta
                Lep1phi = Lep2_phi
                Lep1iso = Lep2_iso
                Lep1id = Lep2_id
                Lep1mom = Lep2_mom

                Lep2pt = Lep3_pt/M2
                Lep2eta = Lep3_eta
                Lep2phi = Lep3_phi
                Lep2iso = Lep3_iso
                Lep2id = Lep3_id
                Lep2mom = Lep3_mom

                dR = deltaR(Lep1eta,Lep1phi,Lep2eta,Lep2phi)

                Selector_alts = np.array((Lep1pt,Lep1eta,Lep1phi,Lep1iso,Lep1id,
                                          Lep2pt,Lep2eta,Lep2phi,Lep2iso,Lep2id,
                                          met, met_phi, dR))
                Selector_alts = normalize_alt(Selector_alts)
                Selector_alts = Selector_alts.reshape(1,-1)
                prediction_2 = ZSelector_alt.predict(Selector_alts)
                
                if prediction_1[0] > prediction_2[0]:
                    nna_guess_slot = 1
                else:
                    nna_guess_slot = 3
                
        elif Lep1_id + Lep3_id == 0:
            M1 = (Lep1 + Lep3).M()
            
            Lep1pt = Lep1_pt/M1
            Lep1eta = Lep1_eta
            Lep1phi = Lep1_phi
            Lep1iso = Lep1_iso
            Lep1id = Lep1_id
            Lep1mom = Lep1_mom

            Lep2pt = Lep3_pt/M1
            Lep2eta = Lep3_eta
            Lep2phi = Lep3_phi
            Lep2iso = Lep3_iso
            Lep2id = Lep3_id
            Lep2mom = Lep3_mom
            
            dR = deltaR(Lep1eta,Lep1phi,Lep2eta,Lep2phi)
            
            Selector_alts = np.array((Lep1pt,Lep1eta,Lep1phi,Lep1iso,Lep1id,
                                      Lep2pt,Lep2eta,Lep2phi,Lep2iso,Lep2id,
                                      met, met_phi, dR))
            Selector_alts = normalize_alt(Selector_alts)
            Selector_alts = Selector_alts.reshape(1,-1)
            prediction_1 = ZSelector_alt.predict(Selector_alts)
            
            if Lep1_id + Lep2_id == 0:
                M2 = (Lep1 + Lep2).M()
                
                Lep1pt = Lep1_pt/M2
                Lep1eta = Lep1_eta
                Lep1phi = Lep1_phi
                Lep1iso = Lep1_iso
                Lep1id = Lep1_id
                Lep1mom = Lep1_mom

                Lep2pt = Lep2_pt/M2
                Lep2eta = Lep2_eta
                Lep2phi = Lep2_phi
                Lep2iso = Lep2_iso
                Lep2id = Lep2_id
                Lep2mom = Lep2_mom

                dR = deltaR(Lep1eta,Lep1phi,Lep2eta,Lep2phi)

                Selector_alts = np.array((Lep1pt,Lep1eta,Lep1phi,Lep1iso,Lep1id,
                                          Lep2pt,Lep2eta,Lep2phi,Lep2iso,Lep2id,
                                          met, met_phi, dR))
                Selector_alts = normalize_alt(Selector_alts)
                Selector_alts = Selector_alts.reshape(1,-1)
                prediction_2 = ZSelector_alt.predict(Selector_alts)
                
                if prediction_1[0] > prediction_2[0]:
                    nna_guess_slot = 2
                else:
                    nna_guess_slot = 1
                
            elif Lep2_id + Lep3_id == 0:
                M2 = (Lep2 + Lep3).M()
                
                Lep1pt = Lep2_pt/M2
                Lep1eta = Lep2_eta
                Lep1phi = Lep2_phi
                Lep1iso = Lep2_iso
                Lep1id = Lep2_id
                Lep1mom = Lep2_mom

                Lep2pt = Lep3_pt/M2
                Lep2eta = Lep3_eta
                Lep2phi = Lep3_phi
                Lep2iso = Lep3_iso
                Lep2id = Lep3_id
                Lep2mom = Lep3_mom

                dR = deltaR(Lep1eta,Lep1phi,Lep2eta,Lep2phi)

                Selector_alts = np.array((Lep1pt,Lep1eta,Lep1phi,Lep1iso,Lep1id,
                                          Lep2pt,Lep2eta,Lep2phi,Lep2iso,Lep2id,
                                          met, met_phi, dR))
                Selector_alts = normalize_alt(Selector_alts)
                Selector_alts = Selector_alts.reshape(1,-1)
                prediction_2 = ZSelector_alt.predict(Selector_alts)
                
                if prediction_1[0] > prediction_2[0]:
                    nna_guess_slot = 2
                else:
                    nna_guess_slot = 3
            
        # ----------------------------------------------------

#         vec = Lep1 + Lep2 + Lep3 + Met
#         Mu3 = Lep1 + Lep2 + Lep3
#         TotMass = vec.M()
#         Mu3Mass = Mu3.M()

        if corr_slot == low_guess_slot:
            low_cor_events+=1
        if corr_slot == hig_guess_slot:
            hig_cor_events+=1
        if corr_slot == nn_guess_slot:
            nn_cor_events+=1
        if corr_slot == nna_guess_slot:
            nna_cor_events+=1

        tot_events+=1

    #print "Mass %i, Percent correct: %f" % (m, cor_events/tot_events*100)
    low_accuracy.append(low_cor_events/tot_events)
    hig_accuracy.append(hig_cor_events/tot_events)
    nn_accuracy.append(nn_cor_events/tot_events)
    nna_accuracy.append(nna_cor_events/tot_events)
```
